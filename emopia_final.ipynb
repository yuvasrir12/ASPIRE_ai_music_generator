```python
import tensorflow as tf
import numpy as np

# Create the LSTM-based model for music generation
def create_music_model(input_shape, condition_shape):
    # Input for note sequences
    inputs = tf.keras.Input(shape=input_shape)

    # Input for emotion condition (one-hot encoded)
    conditions = tf.keras.Input(shape=condition_shape)

    # LSTM layers for sequence generation
    x = tf.keras.layers.LSTM(128, return_sequences=True)(inputs)

    # Repeat the emotion condition across time steps (50 steps)
    repeated_conditions = tf.keras.layers.RepeatVector(input_shape[0])(conditions)

    # Concatenate the LSTM output with the repeated emotion conditions
    combined = tf.keras.layers.Concatenate()([x, repeated_conditions])

    # Output dense layers for generating notes (e.g., pitch, duration)
    pitch_output = tf.keras.layers.TimeDistributed(
        tf.keras.layers.Dense(128, activation='softmax')
    )(combined)  # Now predicting 50 timesteps, each with 128 possible pitch classes

    duration_output = tf.keras.layers.TimeDistributed(
        tf.keras.layers.Dense(1)
    )(combined)  # Now predicting 50 timesteps, each with one duration value

    return tf.keras.Model([inputs, conditions], [pitch_output, duration_output])

# Define input shapes (adjust based on your data)
input_shape = (50, 4)  # e.g., 50 time steps with 4 features (start, end, pitch, velocity)
condition_shape = (4,)  # Assuming 4 emotion classes

# Create the model
model = create_music_model(input_shape, condition_shape)
model.summary()

# Prepare the data (replace this with actual data preparation logic)
X_sequences = np.random.rand(1, 50, 4)  # Input MIDI sequences
y_pitch = np.random.randint(0, 128, size=(1, 50))  # Random pitch for demonstration
y_duration = np.random.rand(1, 50, 1)  # Random durations for demonstration
emotion_inputs = np.array([[0, 0, 0, 1]])  # Example one-hot encoded emotion (Q1)

# One-hot encode the target pitch values to match the model's output
y_pitch_one_hot = tf.keras.utils.to_categorical(y_pitch, num_classes=128)

# Compile the model
model.compile(optimizer='adam', loss=['categorical_crossentropy', 'mean_squared_error'])

# Train the model (this is just a simple example; use actual data)
model.fit([X_sequences, emotion_inputs], [y_pitch_one_hot, y_duration], epochs=30)
```

```python
def generate_music(model, seed_sequence, emotion_input, num_notes=250):
    generated_notes = seed_sequence.copy()  # Start with the seed sequence
    for _ in range(num_notes):
        # Predict the next note
        next_note = model.predict([generated_notes, emotion_input])

        # Extract the pitch predictions and duration predictions
        next_pitch = next_note[0]  # Pitch output
        next_duration = next_note[1]  # Duration output

        # Get the last pitch from next_pitch, which is shaped (1, 128)
        next_pitch = np.argmax(next_pitch[:, -1, :], axis=1)  # Get the index of the highest probability
        next_duration = next_duration[:, -1, :]  # Get the last duration prediction

        # Create a new note based on the last pitch and duration
        new_note = np.array([[0.0, 0.0, next_pitch[0], next_duration[0][0]]])  # [start, end, pitch, velocity]

        # Append the predicted note to the sequence
        generated_notes = np.append(generated_notes[:, 1:, :], new_note.reshape(1, 1, -1), axis=1)  # Shift sequence left and add new note

    return generated_notes
```

```python
# Example: Use a random seed and user-defined emotion input (Q1: Happy)
seed_sequence = np.random.rand(1, 50, 4)  # Example seed sequence
emotion_input = np.array([[0, 0, 0, 1]])  # Example emotion input (Q1: Happy)

# Generate music
generated_music = generate_music(model, seed_sequence, emotion_input)

# Check the shape of the generated music
print("Shape of generated music:", generated_music.shape)
```

```python
import pretty_midi

def notes_to_midi(generated_notes, output_file):
    midi = pretty_midi.PrettyMIDI()
    instrument = pretty_midi.Instrument(program=0)  # Acoustic Grand Piano

    current_time = 0.0  # Initialize current time for sequential note generation

    for note in generated_notes[0]:  # Assuming the format is (1, num_notes, features)
        start, end, pitch, velocity = note

        # If the model-generated start and end times are invalid, fix them
        if start == 0.0 and end == 0.0:
            start = current_time
            end = start + 0.5  # Give a fixed duration of 0.5 seconds
            current_time += 0.5  # Increment time for the next note

        # Scale the velocity to be in the range of 0-127 for MIDI
        velocity = max(0, min(127, int(velocity * 127)))

        # Ensure pitch is within valid range (0-127)
        pitch = max(0, min(127, int(pitch)))

        # Only add the note if the pitch is valid
        if pitch >= 0 and pitch <= 127:
            midi_note = pretty_midi.Note(
                velocity=int(velocity),
                pitch=int(pitch),
                start=float(start),
                end=float(end)
            )
            instrument.notes.append(midi_note)

    # Only write the MIDI file if there are valid notes
    if len(instrument.notes) > 0:
        midi.instruments.append(instrument)
        midi.write(output_file)
        print(f"MIDI file saved as {output_file}.")
    else:
        print("No valid notes to write to the MIDI file.")

# Example of saving the generated music to a MIDI file
notes_to_midi(generated_music, 'generated_music_fixed.mid')

# Example of checking the generated notes for valid pitch and velocity
for note in generated_music[0]:
    start, end, pitch, velocity = note
    print(f"Start: {start}, End: {end}, Pitch: {pitch}, Velocity: {velocity}")
```
